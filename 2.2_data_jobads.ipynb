# JOB OFFER'S DATA

```python
# Load nessesary libraries.
import selenium 
import time
from datetime import datetime
import pandas as pd
from time import sleep
from random import randint
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException
```

```python
# Version check.
print('Selenium version used:', selenium.__version__)
```

```python
# Start up the driver with with Google Chrome using incognito mode and define job with location of interest.
option= webdriver.ChromeOptions()
option.add_argument("--incognito")

jobs = ['registered+nurse', 'electrician', 'data+analyst']

pagination_url = 'https://ie.indeed.com/jobs?q={}&l=Dublin%2C+County+Dublin&radius=25&filter=0&sort=date&start={}'

current_date = datetime.now().date().strftime('%B %d, %Y')
```

### 1. Web Scraping
**registered nurse ads**

```python
# Determine how many jobs are open for a specific position and how many pages to loop through.
start = time.time()
job = jobs[0]

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=option)
driver.get(pagination_url.format(job, 0))

sleep(randint(4, 9))
job_number = driver.find_element(By.CLASS_NAME,'jobsearch-JobCountAndSortPane-jobCount').text
max_iter_pgs=int(job_number.split(' ')[0])//15 

driver.quit() 
end = time.time()

# Check results.
print('REGISTERED NURSE')
print(f'Total number of vacancies available in Dublin area on {current_date}: {job_number}.')
print('Maximum number of iterable pages for the search:', max_iter_pgs, 'pages')
print('\n')
print('Action was completed in:', end-start, 'seconds.')
```

```python
# Create a list of each job advertisement’s title, id, posted date and the link to access it.
start = time.time()
job = jobs[0]
nurse_joblist = []

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=option)
sleep(randint(4, 9))

for i in range(0, max_iter_pgs):
    driver.get(pagination_url.format(job, i*10))
    
    sleep(randint(4, 9))

    job_page = driver.find_element(By.ID, 'mosaic-jobResults')
    job_posts = job_page.find_elements(By.CLASS_NAME, 'job_seen_beacon')

    for job_post in job_posts:
        job_title = job_post.find_element(By.CLASS_NAME, 'jobTitle') #print(job_title.text)
        job_id = job_title.find_element(By.CSS_SELECTOR, 'a').get_attribute("id") #print(job_id)
        job_link = job_title.find_element(By.CSS_SELECTOR, 'a').get_attribute("href") #print(job_link)        

        try:
            job_date = job_post.find_element(By.CLASS_NAME, 'date').text
        except Exception as e:
            job_date = 'not available'

        nurse_joblist.append([job_title.text, job_id, job_link, job_date])

driver.quit() 
end = time.time()

# Check results.
print('REGISTERED NURSE JOB ADS')
print('The total count of scraped job vacancies:', len(nurse_joblist), 'jobs.\n')
for x in range(2):
    print(f'JOB ADD NO.{x+1}:')
    print(nurse_joblist[x][0])
    print(nurse_joblist[x][1])
    print(nurse_joblist[x][2])
    print(nurse_joblist[x][3])
print('\n')

print(f'The extraction was completed in: {(end-start)//60} minutes and {(end-start)%60} seconds.')
```
